```python
import cv2
import matplotlib.pyplot as plt

# 1. Resmi Okuma (BGR formatında)
para1_bgr = cv2.imread("para1.jpg")
plt.imshow(cv2.cvtColor(para1_bgr, cv2.COLOR_BGR2RGB))
plt.title("1. Orijinal Resim (BGR Formatında)")
plt.axis("off")
plt.show()

# 2. BGR'den RGB'ye çevirme
para1_rgb = cv2.cvtColor(para1_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(para1_rgb)
plt.title("2. RGB Formatına Çevrilmiş Resim")
plt.axis("off")
plt.show()

# 3. RGB'den Gri Tonlamaya Çevirme
para1_gri = cv2.cvtColor(para1_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(para1_gri, cmap="gray")
plt.title("3. Gri Tonlamaya Çevrilmiş Resim")
plt.axis("off")
plt.show()

# 4. Gauss Bulanıklaştırma
bulanik_para1 = cv2.GaussianBlur(para1_gri, (7, 7), 0)
plt.imshow(bulanik_para1, cmap="gray")
plt.title("4. Gauss Bulanıklaştırma")
plt.axis("off")
plt.show()

# 5. Otsu Threshold (Eşikleme Uygulaması)
_, binary_para1 = cv2.threshold(bulanik_para1, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(binary_para1, cmap="gray")
plt.title("5. Otsu Threshold (Eşikleme)")
plt.axis("off")
plt.show()

# 6. Canny Kenar Algılama
kenar_para1 = cv2.Canny(bulanik_para1, 100, 200)
plt.imshow(kenar_para1, cmap="gray")
plt.title("6. Canny Kenar Algılama")
plt.axis("off")
plt.show()

# 7. Kontur Tespiti
contours, hierarchy = cv2.findContours(kenar_para1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 8. Kontur Alanlarına Göre Sıralama
contours, hierarchy = cv2.findContours(kenar_para1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
para1_rgb = cv2.cvtColor(para1_bgr, cv2.COLOR_BGR2RGB)
cv2.drawContours(para1_rgb, contours, -1, (255, 0, 0), 2)
min_area = 35 # Filtreleme için minimum alan eşiği
sorted_contours = sorted(contours, key=cv2.contourArea, reverse=False)  # Büyükten küçüğe sırala
filtered_contours = [kontor for kontor in sorted_contours if cv2.contourArea(kontor) > min_area]

# 9. Büyük konturların etrafına dikdörtgen çizme ve numaralandırma
para1_rgb_numaralandirilmis = para1_rgb.copy()

for i, kontor in enumerate(filtered_contours):
    x, y, w, h = cv2.boundingRect(kontor)  # Konturu çevreleyen dikdörtgeni bul
    cv2.rectangle(para1_rgb_numaralandirilmis, (x, y), (x + w, y + h), (0, 255, 0), 3)  # Yeşil dikdörtgen çiz
    cv2.putText(para1_rgb_numaralandirilmis, str(i + 1), (x, y - 10), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 255), 2)  # Numaralandır
     # Dinamik font ölçekleme
    font_scale = min(w, h) / 150  # Alanın büyüklüğüne göre font ölçeği
    font_scale = max(font_scale, 0.5)  # Minimum font boyutu
    text_x = x + 0  # X ekseninde içe kaydırma
    text_y = y + h // 2  # Y ekseninde ortalama

    cv2.putText(para1_rgb_numaralandirilmis, "Veli Ayata", (text_x, text_y),
                cv2.FONT_HERSHEY_SIMPLEX, font_scale, (0, 0, 250), 1)  # Mor renk ile yazı ekle
    print(f"Kontur {i + 1} alanı: {cv2.contourArea(kontor)} piksel")  # Alanı yazdır

# 10. Sonucu Gösterme
plt.imshow(para1_rgb_numaralandirilmis)
plt.title("7. Kontur Tespiti ve Çizimi - Paraların Alanına Göre Sıralama")
plt.axis("off")
plt.show()
```
